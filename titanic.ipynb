### In this challenge, we ask you to build a predictive model that answers the question: “what sorts of people were more likely to survive?” using passenger data (ie name, age, gender, socio-economic class, etc)

# Import the Libraries

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importing The Dataset

```python
df1 = pd.read_csv('train.csv')
df2 = pd.read_csv('test.csv')
df1
```

```python
cols = df1.columns 
cols
```

```python
df1.columns.get_loc("Embarked") 
```

```python
z1 = df1['Survived']
```

```python
df1 = df1.drop('Survived',1)
df1
```

```python
df1['Survived']= z1 #df = df[['PassengerId', 'Pclass', 'Name', 'Sex', 'Age', 'SibSp','Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked', 'Survived']
df1
```

```python
df1.index=df1['PassengerId']
df1
```

```python
cols2 = df2.columns 
cols2
```

```python
df2.index=df2['    PassengerId']
df2
```

```python
df1 = df1.drop(['PassengerId','Name','Ticket', 'Fare', 'Cabin'],axis=1)
df1
```

```python

df2 = df2.drop(['Name','Ticket','Fare','    PassengerId','Cabin'],axis=1)
df2
```

```python
x_train = df1.iloc[:, :-1].values# matrix of features
y_train = df1.iloc[:, -1].values#dependent variable
x_test = df2.iloc[:].values# matrix of features
```

# Importing Missing Data

```python
#For Numerical Data
from sklearn.impute import SimpleImputer
imputer1 = SimpleImputer(missing_values=np.nan,strategy='median')
imputer2 = SimpleImputer(missing_values=np.nan,strategy='median')
imputer1.fit(x_train[: ,2:3])#to connect our imputer object to the matrix of features X
x_train[: ,2:3]=imputer1.transform(x_train[: ,2:3])# to apply the transformation
imputer2.fit(x_test[: ,2:3])
x_test[: ,2:3]=imputer2.transform(x_test[: ,2:3])
```

```python
#For Categorical Data
imputer3 = SimpleImputer(missing_values=np.nan,strategy='most_frequent')
imputer3.fit(x_train[: ,5:6])
x_train[: ,5:6]=imputer3.transform(x_train[: ,5:6])# to apply the transformation
x_train
```

```python
x_test
```

```python
#For Categorical Data
imputer4 = SimpleImputer(missing_values=np.nan,strategy='most_frequent')
imputer4.fit(x_test[: ,5:6])
x_test[: ,5:6]=imputer4.transform(x_test[: ,5:6])# to apply the transformation
x_test
```

## Encoding Independent variable or Categorical Data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [('encoder',OneHotEncoder(),[5])],remainder='passthrough')
x_train=np.array(ct.fit_transform(x_train))
x_test=np.array(ct.fit_transform(x_test))
```

```python
x_train[2]
```

```python
x_test[2]
```

## Encoding Dependent variable or Ordinal Data


```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
x_train[:,4]=le.fit_transform(x_train[:,4])
x_test[:,4]=le.fit_transform(x_test[:,4])
x_test
```

```python
x_test
```

```python
np.set_printoptions(threshold=np.inf)
x_train
```

## Feature Scaling

```python
# Feature Scaling:Standardisation
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)   
x_test=sc.transform(x_test) 
x_train
```

# Building the ANN

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Initializing the ANN
ann= Sequential()
# Adding the input layer and the first hidden layer
ann.add(Dense(units=100, activation='relu'))
#Adding the 2nd hidden layer
ann.add(Dense(units=100, activation='relu'))
#Adding the output layer
ann.add(Dense(units=1, activation='sigmoid'))
```

# TRAINING THE ANN

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(x_train,y_train,batch_size=32 , epochs=100,verbose =2)
```

# Predicting the test set results

```python
y_pred = ann.predict(x_test)
y_pred = (y_pred>.5)
y_pred
```
